````python
# â° æ™‚é–“å¸¯åˆ¥éœ€è¦ãƒ‘ã‚¿ãƒ¼ãƒ³åˆ†æã‚¨ãƒ³ã‚¸ãƒ³

print("\nâ° æ™‚é–“å¸¯åˆ¥éœ€è¦ãƒ‘ã‚¿ãƒ¼ãƒ³åˆ†æ\n")
print("="*120)

if not HOURLY_DATA_AVAILABLE:
    print("\nğŸ“Š æ“¬ä¼¼æ™‚é–“å¸¯ãƒ‡ãƒ¼ã‚¿ã‚’ç”Ÿæˆã—ã¾ã™...")
    
    # æ—¥åˆ¥ãƒ‡ãƒ¼ã‚¿ã‹ã‚‰æ™‚é–“å¸¯ãƒ‡ãƒ¼ã‚¿ã‚’æ“¬ä¼¼ç”Ÿæˆ
    # å®Ÿéš›ã®ãƒ“ã‚¸ãƒã‚¹ãƒ‘ã‚¿ãƒ¼ãƒ³ã«åŸºã¥ãæ™‚é–“å¸¯åˆ†å¸ƒ
    hourly_patterns = {
        0: 0.005, 1: 0.003, 2: 0.002, 3: 0.001, 4: 0.002, 5: 0.005,
        6: 0.015, 7: 0.045, 8: 0.070, 9: 0.055, 10: 0.045, 11: 0.050,
        12: 0.090, 13: 0.075, 14: 0.055, 15: 0.045, 16: 0.050, 17: 0.065,
        18: 0.095, 19: 0.085, 20: 0.070, 21: 0.055, 22: 0.040, 23: 0.020
    }
    
    hourly_data_list = []
    
    for _, day_row in my_df.iterrows():
        date = day_row['æ—¥ä»˜']
        daily_sales = day_row['å£²ä¸Šé‡‘é¡']
        
        # æ›œæ—¥ã«ã‚ˆã‚‹èª¿æ•´
        weekday = date.weekday()
        weekday_factor = 1.0
        if weekday >= 5:  # åœŸæ—¥
            weekday_factor = 1.15
        
        for hour in range(24):
            base_ratio = hourly_patterns[hour] * weekday_factor
            # ãƒ©ãƒ³ãƒ€ãƒ ãƒã‚¤ã‚ºã‚’è¿½åŠ ï¼ˆÂ±10%ï¼‰
            noise = np.random.uniform(0.9, 1.1)
            hour_sales = daily_sales * base_ratio * noise
            
            hourly_data_list.append({
                'æ—¥ä»˜': date,
                'åº—èˆ—': MY_STORE,
                'æ™‚é–“å¸¯': hour,
                'å£²ä¸Šé‡‘é¡': hour_sales,
                'æ›œæ—¥': weekday
            })
    
    df_hourly = pd.DataFrame(hourly_data_list)
    print(f"   ç”Ÿæˆå®Œäº†: {len(df_hourly):,}ãƒ¬ã‚³ãƒ¼ãƒ‰")

# æ™‚é–“å¸¯åˆ¥ã®é›†è¨ˆ
hourly_summary = df_hourly.groupby('æ™‚é–“å¸¯').agg({
    'å£²ä¸Šé‡‘é¡': ['sum', 'mean', 'std', 'count']
}).reset_index()

hourly_summary.columns = ['æ™‚é–“å¸¯', 'ç·å£²ä¸Š', 'å¹³å‡å£²ä¸Š', 'æ¨™æº–åå·®', 'ãƒ‡ãƒ¼ã‚¿æ—¥æ•°']
hourly_summary['å£²ä¸Šæ¯”ç‡'] = hourly_summary['ç·å£²ä¸Š'] / hourly_summary['ç·å£²ä¸Š'].sum() * 100
hourly_summary['å¤‰å‹•ä¿‚æ•°'] = hourly_summary['æ¨™æº–åå·®'] / hourly_summary['å¹³å‡å£²ä¸Š']

# ãƒ”ãƒ¼ã‚¯ã‚¿ã‚¤ãƒ ã®ç‰¹å®š
peak_hours = hourly_summary.nlargest(3, 'å¹³å‡å£²ä¸Š')

print(f"\nğŸ“Š æ™‚é–“å¸¯åˆ¥å£²ä¸Šã‚µãƒãƒªãƒ¼")
print("="*120)
print(f"   ç·å£²ä¸Š: Â¥{hourly_summary['ç·å£²ä¸Š'].sum():,.0f}")
print(f"   å¹³å‡æ™‚é–“å£²ä¸Š: Â¥{hourly_summary['å¹³å‡å£²ä¸Š'].mean():,.0f}")
print(f"   æœ€å¤§æ™‚é–“å£²ä¸Š: Â¥{hourly_summary['å¹³å‡å£²ä¸Š'].max():,.0f} ({hourly_summary.loc[hourly_summary['å¹³å‡å£²ä¸Š'].idxmax(), 'æ™‚é–“å¸¯']:.0f}æ™‚å°)")
print(f"   æœ€å°æ™‚é–“å£²ä¸Š: Â¥{hourly_summary['å¹³å‡å£²ä¸Š'].min():,.0f} ({hourly_summary.loc[hourly_summary['å¹³å‡å£²ä¸Š'].idxmin(), 'æ™‚é–“å¸¯']:.0f}æ™‚å°)")

print(f"\n\nâ­ ãƒ”ãƒ¼ã‚¯ã‚¿ã‚¤ãƒ  TOP 3")
print("="*120)
print(f"\n{'é †ä½':<6} {'æ™‚é–“å¸¯':<10} {'å¹³å‡å£²ä¸Š':>15} {'å£²ä¸Šæ¯”ç‡':>10} {'å¤‰å‹•ä¿‚æ•°':>12} {'æ¨å¥¨å¯¾ç­–'}")
print("="*120)

for rank, (_, row) in enumerate(peak_hours.iterrows(), 1):
    hour = int(row['æ™‚é–“å¸¯'])
    avg_sales = f"Â¥{row['å¹³å‡å£²ä¸Š']:,.0f}"
    ratio = f"{row['å£²ä¸Šæ¯”ç‡']:.1f}%"
    cv = f"{row['å¤‰å‹•ä¿‚æ•°']:.3f}"
    
    # æ™‚é–“å¸¯ã«å¿œã˜ãŸå¯¾ç­–
    if 7 <= hour <= 9:
        action = "æœé£Ÿå•†å“ã‚’å¼·åŒ–"
    elif 11 <= hour <= 13:
        action = "æ˜¼é£Ÿå•†å“ã‚’å¼·åŒ–"
    elif 17 <= hour <= 20:
        action = "å¤•é£Ÿãƒ»æƒ£èœã‚’å¼·åŒ–"
    else:
        pass
        action = "ã‚¹ã‚¿ãƒƒãƒ•å¢—å“¡"
    
    print(f"{rank:<6} {hour:02d}:00-{hour:02d}:59  {avg_sales:>15} {ratio:>10} {cv:>12} {action}")

print(f"\n\nğŸ’¡ ãƒ”ãƒ¼ã‚¯ã‚¿ã‚¤ãƒ æœ€é©åŒ–ã®æ¨å¥¨ã‚¢ã‚¯ã‚·ãƒ§ãƒ³")
print("="*120)

for rank, (_, row) in enumerate(peak_hours.iterrows(), 1):
    hour = int(row['æ™‚é–“å¸¯'])
    print(f"\nã€ãƒ”ãƒ¼ã‚¯{rank}ã€‘ {hour:02d}:00-{hour:02d}:59 (å£²ä¸Šæ¯”ç‡ {row['å£²ä¸Šæ¯”ç‡']:.1f}%)")
    print("-"*120)
    
    print(f"   ğŸ“‹ ç¾çŠ¶:")
    print(f"      å¹³å‡å£²ä¸Š: Â¥{row['å¹³å‡å£²ä¸Š']:,.0f}")
    print(f"      å¤‰å‹•ä¿‚æ•°: {row['å¤‰å‹•ä¿‚æ•°']:.3f} ({'é«˜å¤‰å‹•' if row['å¤‰å‹•ä¿‚æ•°'] > 0.5 else 'å®‰å®š'})")
    
    print(f"\n   ğŸ¯ æ¨å¥¨å¯¾ç­–:")
    
    if 7 <= hour <= 9:
        print(f"      1. æœé£Ÿå•†å“ï¼ˆãŠã«ãã‚Šã€ã‚µãƒ³ãƒ‰ã‚¤ãƒƒãƒã€ã‚³ãƒ¼ãƒ’ãƒ¼ï¼‰ã®åœ¨åº«ã‚’+50%å¢—é‡")
        print(f"      2. ãƒ¬ã‚¸ã‚¹ã‚¿ãƒƒãƒ•ã‚’é€šå¸¸+2åé…ç½®")
        print(f"      3. æœé£Ÿã‚»ãƒƒãƒˆï¼ˆãŠã«ãã‚Š+é£²ã¿ç‰©ï¼‰ã§Â¥50å‰²å¼•ã‚­ãƒ£ãƒ³ãƒšãƒ¼ãƒ³")
        print(f"      4. å‰æ—¥21æ™‚ã¾ã§ã«æœé£Ÿå•†å“ã®è£œå……å®Œäº†")
    elif 11 <= hour <= 13:
        print(f"      1. å¼å½“ãƒ»æƒ£èœã®åœ¨åº«ã‚’+70%å¢—é‡")
        print(f"      2. ãƒ¬ã‚¸ã‚¹ã‚¿ãƒƒãƒ•ã‚’é€šå¸¸+3åé…ç½®")
        print(f"      3. å¼å½“æ—©å‰²ï¼ˆ11æ™‚å°ï¼‰10%OFFã§å‰å€’ã—è²©å£²")
        print(f"      4. é›»å­ãƒ¬ãƒ³ã‚¸ã‚’å¢—è¨­ã—ã¦ãƒ¬ã‚¸å¾…ã¡æ™‚é–“ã‚’çŸ­ç¸®")
    elif 17 <= hour <= 20:
        print(f"      1. å¤•é£Ÿãƒ»æƒ£èœã®åœ¨åº«ã‚’+60%å¢—é‡")
        print(f"      2. ãƒ¬ã‚¸ã‚¹ã‚¿ãƒƒãƒ•ã‚’é€šå¸¸+2åé…ç½®")
        print(f"      3. è¦‹åˆ‡ã‚Šå‰ã®ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³ï¼ˆ18æ™‚å°10%OFFï¼‰")
        print(f"      4. ãƒ“ãƒ¼ãƒ«ãƒ»ã¤ã¾ã¿ã®ã‚¨ãƒ³ãƒ‰é™³åˆ—")
    else:
        pass
        print(f"      1. ã‚¹ã‚¿ãƒƒãƒ•ã‚’é€šå¸¸+1-2åé…ç½®")
        print(f"      2. åœ¨åº«åˆ‡ã‚Œé˜²æ­¢ã®ãŸã‚ã®è¿½åŠ ç™ºæ³¨")
        print(f"      3. ãƒ¬ã‚¸åŠ¹ç‡åŒ–ï¼ˆã‚»ãƒ«ãƒ•ãƒ¬ã‚¸èª˜å°ï¼‰")
    
    # ROIè©¦ç®—
    additional_staff_cost = 1200 * 2  # æ™‚çµ¦Â¥1,200 Ã— 2å
    inventory_cost = row['å¹³å‡å£²ä¸Š'] * 0.5 * 0.3  # åœ¨åº«å¢—50% Ã— åŸä¾¡ç‡30%
    expected_sales_increase = row['å¹³å‡å£²ä¸Š'] * 0.15  # 15%å£²ä¸Šå‘ä¸Š
    expected_profit = expected_sales_increase * 0.3  # ç²—åˆ©ç‡30%
    
    total_cost = additional_staff_cost + inventory_cost
    net_profit = expected_profit - total_cost
    roi = (net_profit / total_cost) * 100 if total_cost > 0 else 0
    
    print(f"\n   ğŸ’° ROIè©¦ç®—:")
    print(f"      æŠ•è³‡é¡:   Â¥{total_cost:,.0f} (äººä»¶è²»Â¥{additional_staff_cost:,.0f} + åœ¨åº«Â¥{inventory_cost:,.0f})")
    print(f"      æœŸå¾…åŠ¹æœ: Â¥{expected_profit:,.0f} (å£²ä¸Š+15%æƒ³å®šã€ç²—åˆ©ç‡30%)")
    print(f"      ç´”åˆ©ç›Š:   Â¥{net_profit:,.0f}")
    print(f"      ROI:      {roi:+.0f}%")
````

````python
# ğŸ“£ ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³åŠ¹æœæ¸¬å®šã‚¨ãƒ³ã‚¸ãƒ³

print("\nğŸ“£ ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³åŠ¹æœæ¸¬å®šãƒ»ROIåˆ†æ\n")
print("="*120)

# ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³æœŸé–“ã®å®šç¾©ï¼ˆä¾‹: 8æœˆ1-7æ—¥ã€8æœˆ15-21æ—¥ã‚’ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³æœŸé–“ã¨ã™ã‚‹ï¼‰
promotions = [
    {
        'name': 'å¤ç¥­ã‚Šã‚­ãƒ£ãƒ³ãƒšãƒ¼ãƒ³',
        'start': '2025-08-01',
        'end': '2025-08-07',
        'cost': 50000,  # ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³ã‚³ã‚¹ãƒˆï¼ˆåºƒå‘Šã€å€¤å¼•ãç­‰ï¼‰
        'type': 'å€¤å¼•ã',
        'discount_rate': 0.10  # 10%OFF
    },
    {
        'name': 'ãŠç›†ã‚»ãƒ¼ãƒ«',
        'start': '2025-08-15',
        'end': '2025-08-21',
        'cost': 80000,
        'type': 'ãƒã‚¤ãƒ³ãƒˆ2å€',
        'discount_rate': 0.05  # å®Ÿè³ª5%é‚„å…ƒ
    }
]

print("\nğŸ“Š ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³åŠ¹æœåˆ†æ")
print("="*120)

for promo in promotions:
    promo_start = pd.to_datetime(promo['start'])
    promo_end = pd.to_datetime(promo['end'])
    promo_days = (promo_end - promo_start).days + 1
    
    # ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³æœŸé–“ã®ãƒ‡ãƒ¼ã‚¿
    promo_data = my_df[(my_df['æ—¥ä»˜'] >= promo_start) & (my_df['æ—¥ä»˜'] <= promo_end)]
    
    if len(promo_data) == 0:
        print(f"\nâš ï¸ {promo['name']}: ãƒ‡ãƒ¼ã‚¿ãªã—ï¼ˆæœŸé–“å¤–ï¼‰")
        continue
    
    # BeforeæœŸé–“ï¼ˆãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³å‰ã®åŒæ—¥æ•°ï¼‰
    before_start = promo_start - timedelta(days=promo_days*2)
    before_end = promo_start - timedelta(days=1)
    before_data = my_df[(my_df['æ—¥ä»˜'] >= before_start) & (my_df['æ—¥ä»˜'] <= before_end)]
    
    # AfteræœŸé–“ï¼ˆãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³å¾Œã®åŒæ—¥æ•°ï¼‰
    after_start = promo_end + timedelta(days=1)
    after_end = promo_end + timedelta(days=promo_days)
    after_data = my_df[(my_df['æ—¥ä»˜'] >= after_start) & (my_df['æ—¥ä»˜'] <= after_end)]
    
    # æ—¥æ¬¡é›†è¨ˆ
    promo_daily_sales = promo_data.groupby('æ—¥ä»˜')['å£²ä¸Šé‡‘é¡'].sum()
    before_daily_sales = before_data.groupby('æ—¥ä»˜')['å£²ä¸Šé‡‘é¡'].sum()
    after_daily_sales = after_data.groupby('æ—¥ä»˜')['å£²ä¸Šé‡‘é¡'].sum()
    
    # çµ±è¨ˆé‡
    promo_avg = promo_daily_sales.mean()
    before_avg = before_daily_sales.mean()
    after_avg = after_daily_sales.mean() if len(after_daily_sales) > 0 else 0
    
    # ãƒªãƒ•ãƒˆç‡
    lift_rate = (promo_avg - before_avg) / before_avg * 100 if before_avg > 0 else 0
    
    # tæ¤œå®šï¼ˆãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³æœŸé–“ vs BeforeæœŸé–“ï¼‰
    if len(before_daily_sales) >= 2 and len(promo_daily_sales) >= 2:
        t_stat, p_value = stats.ttest_ind(promo_daily_sales, before_daily_sales)
        is_significant = p_value < 0.05
    else:
        pass
        t_stat, p_value = 0, 1.0
        is_significant = False
    
    # ROIè¨ˆç®—
    total_promo_sales = promo_daily_sales.sum()
    baseline_sales = before_avg * promo_days
    incremental_sales = total_promo_sales - baseline_sales
    
    # ç²—åˆ©ç‡30%æƒ³å®šã€å€¤å¼•ãåˆ†ã‚’å·®ã—å¼•ã
    gross_margin = 0.30
    discount_cost = total_promo_sales * promo['discount_rate']
    incremental_profit = incremental_sales * gross_margin - discount_cost
    net_profit = incremental_profit - promo['cost']
    roi = (net_profit / promo['cost']) * 100 if promo['cost'] > 0 else 0
    
    print(f"\n\n{'='*120}")
    print(f"ã€{promo['name']}ã€‘ ({promo['start']} ~ {promo['end']})")
    print(f"{'='*120}")
    
    print(f"\nğŸ“Š å£²ä¸Šæ¯”è¼ƒ:")
    print(f"   BeforeæœŸé–“å¹³å‡: Â¥{before_avg:,.0f}/æ—¥ ({before_start.strftime('%Y-%m-%d')} ~ {before_end.strftime('%Y-%m-%d')})")
    print(f"   ãƒ—ãƒ­ãƒ¢æœŸé–“å¹³å‡: Â¥{promo_avg:,.0f}/æ—¥ ({promo_start.strftime('%Y-%m-%d')} ~ {promo_end.strftime('%Y-%m-%d')})")
    if len(after_daily_sales) > 0:
        print(f"   AfteræœŸé–“å¹³å‡:  Â¥{after_avg:,.0f}/æ—¥ ({after_start.strftime('%Y-%m-%d')} ~ {after_end.strftime('%Y-%m-%d')})")
    
    print(f"\nğŸ“ˆ åŠ¹æœæŒ‡æ¨™:")
    print(f"   ãƒªãƒ•ãƒˆç‡:       {lift_rate:+.1f}%")
    print(f"   å¢—åˆ†å£²ä¸Š:       Â¥{incremental_sales:,.0f} ({promo_days}æ—¥é–“åˆè¨ˆ)")
    print(f"   çµ±è¨ˆçš„æœ‰æ„æ€§:   {'âœ… æœ‰æ„ (p={p_value:.4f})' if is_significant else 'âŒ éæœ‰æ„ (p={p_value:.4f})'}")
    
    print(f"\nğŸ’° ROIåˆ†æ:")
    print(f"   ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³ã‚³ã‚¹ãƒˆ: Â¥{promo['cost']:,.0f}")
    print(f"   å€¤å¼•ããƒ»é‚„å…ƒã‚³ã‚¹ãƒˆ:   Â¥{discount_cost:,.0f} ({promo['type']})")
    print(f"   å¢—åˆ†ç²—åˆ©:             Â¥{incremental_profit:,.0f} (ç²—åˆ©ç‡30%æƒ³å®š)")
    print(f"   ç´”åˆ©ç›Š:               Â¥{net_profit:,.0f}")
    print(f"   ROI:                  {roi:+.1f}%")
    
    print(f"\nğŸ’¡ è©•ä¾¡ãƒ»æ¨å¥¨:")
    if roi > 100:
        print(f"   ğŸ‰ å¤§æˆåŠŸï¼ ROI {roi:.0f}%ã¯éå¸¸ã«å„ªç§€ã§ã™")
        print(f"      â†’ åŒæ§˜ã®ã‚­ãƒ£ãƒ³ãƒšãƒ¼ãƒ³ã‚’å®šæœŸçš„ã«å®Ÿæ–½ã™ã¹ã")
        print(f"      â†’ æˆåŠŸè¦å› ã‚’åˆ†æã—ã€ä»–ã®æ–½ç­–ã«ã‚‚å¿œç”¨")
    elif roi > 50:
        print(f"   âœ… æˆåŠŸï¼ ROI {roi:.0f}%ã¯è‰¯å¥½ã§ã™")
        print(f"      â†’ ç¶™ç¶šå®Ÿæ–½ã‚’æ¨å¥¨")
        print(f"      â†’ ã‚³ã‚¹ãƒˆå‰Šæ¸›ã®ä½™åœ°ã‚’æ¤œè¨")
    elif roi > 0:
        print(f"   ğŸ“Š ã‚„ã‚„æˆåŠŸã€‚ROI {roi:.0f}%ã¯é»’å­—ã ãŒæ”¹å–„ä½™åœ°ã‚ã‚Š")
        print(f"      â†’ å®Ÿæ–½æ–¹æ³•ã®è¦‹ç›´ã—ã‚’æ¤œè¨")
        print(f"      â†’ ã‚¿ãƒ¼ã‚²ãƒƒãƒˆå•†å“ãƒ»é¡§å®¢ã®çµã‚Šè¾¼ã¿")
    else:
        pass
        print(f"   âš ï¸ å¤±æ•—ã€‚ROI {roi:.0f}%ã¯èµ¤å­—ã§ã™")
        print(f"      â†’ ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³å†…å®¹ã®æŠœæœ¬çš„è¦‹ç›´ã—ãŒå¿…è¦")
        print(f"      â†’ ã‚³ã‚¹ãƒˆæ§‹é€ ã®æ”¹å–„ã¾ãŸã¯å®Ÿæ–½ä¸­æ­¢ã‚’æ¤œè¨")
    
    if not is_significant:
        print(f"\n   âš ï¸ æ³¨æ„: çµ±è¨ˆçš„ã«æœ‰æ„ãªåŠ¹æœãŒèªã‚ã‚‰ã‚Œã¾ã›ã‚“ã§ã—ãŸï¼ˆp={p_value:.4f})")
        print(f"      â†’ å£²ä¸Šå¤‰å‹•ãŒå¶ç„¶ã®å¯èƒ½æ€§ãŒã‚ã‚Šã¾ã™")
        print(f"      â†’ ã‚ˆã‚Šé•·æœŸé–“ã¾ãŸã¯å¤§è¦æ¨¡ãªãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³ã§å†æ¤œè¨¼ã‚’æ¨å¥¨")

print(f"\n\nâœ… ãƒ—ãƒ­ãƒ¢ãƒ¼ã‚·ãƒ§ãƒ³åŠ¹æœæ¸¬å®šå®Œäº†")
````

````python
# ğŸ”® What-ifã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ã‚¨ãƒ³ã‚¸ãƒ³

print("\nğŸ”® What-ifã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ãƒ»æ–½ç­–äº‹å‰è©•ä¾¡\n")
print("="*120)

# ç¾çŠ¶ã®ãƒ™ãƒ¼ã‚¹ãƒ©ã‚¤ãƒ³
baseline_daily_sales = my_df.groupby('æ—¥ä»˜')['å£²ä¸Šé‡‘é¡'].sum().mean()
baseline_daily_quantity = my_df.groupby('æ—¥ä»˜')['å£²ä¸Šæ•°é‡'].sum().mean()
baseline_unit_price = baseline_daily_sales / baseline_daily_quantity if baseline_daily_quantity > 0 else 0

print(f"\nğŸ“Š ç¾çŠ¶ãƒ™ãƒ¼ã‚¹ãƒ©ã‚¤ãƒ³")
print("="*120)
print(f"   å¹³å‡æ—¥å•†:   Â¥{baseline_daily_sales:,.0f}")
print(f"   å¹³å‡å®¢æ•°:   {baseline_daily_quantity:.0f}å€‹/æ—¥")
print(f"   å¹³å‡å˜ä¾¡:   Â¥{baseline_unit_price:.0f}")

# ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ã‚·ãƒŠãƒªã‚ª
scenarios = [
    {
        'name': 'å…¨å•†å“5%å€¤ä¸‹ã’',
        'price_change': -0.05,
        'quantity_elasticity': 0.15,  # ä¾¡æ ¼å¼¾åŠ›æ€§: 5%å€¤ä¸‹ã’ã§æ•°é‡15%å¢—
        'cost': baseline_daily_sales * 0.05 * 30,  # å€¤ä¸‹ã’åˆ†Ã—30æ—¥
        'description': 'ç«¶åˆå¯¾æŠ—ã®ãŸã‚å…¨å•†å“ã‚’5%å€¤ä¸‹ã’'
    },
    {
        'name': 'äººæ°—å•†å“10%å€¤ä¸‹ã’',
        'price_change': -0.10,
        'quantity_elasticity': 0.25,  # äººæ°—å•†å“ã¯ä¾¡æ ¼ã«æ•æ„Ÿ
        'affected_sales_ratio': 0.30,  # å£²ä¸Šã®30%ãŒå¯¾è±¡
        'cost': baseline_daily_sales * 0.30 * 0.10 * 30,
        'description': 'å£²ä¸ŠTOP 30%ã®å•†å“ã‚’10%å€¤ä¸‹ã’'
    },
    {
        'name': 'ãƒ”ãƒ¼ã‚¯ã‚¿ã‚¤ãƒ ã‚¹ã‚¿ãƒƒãƒ•+2å',
        'staff_increase': 2,
        'hours_per_day': 3,  # 3æ™‚é–“Ã—30æ—¥
        'hourly_wage': 1200,
        'sales_increase': 0.10,  # ãƒ¬ã‚¸å¾…ã¡è§£æ¶ˆã§10%å£²ä¸Šå¢—
        'cost': 2 * 3 * 1200 * 30,
        'description': 'ãƒ”ãƒ¼ã‚¯æ™‚é–“å¸¯(3æ™‚é–“)ã«ã‚¹ã‚¿ãƒƒãƒ•2åå¢—å“¡'
    },
    {
        'name': 'åœ¨åº«20%å¢—é‡',
        'inventory_increase': 0.20,
        'stockout_reduction': 0.50,  # æ¬ å“ç‡50%å‰Šæ¸›
        'opportunity_loss_current': 50000,  # ç¾åœ¨ã®æœˆé–“æ¬ å“æå¤±
        'cost': baseline_daily_sales * 0.20 * 0.30,  # åœ¨åº«å¢—Ã—åŸä¾¡ç‡30%
        'description': 'å…¨å•†å“ã®åœ¨åº«ã‚’20%å¢—é‡ã—ã¦æ¬ å“å‰Šæ¸›'
    },
    {
        'name': 'æ–°å•†å“50SKUå°å…¥',
        'new_products': 50,
        'expected_sales_per_sku': 500,  # 1SKUã‚ãŸã‚Šæœˆé–“å£²ä¸Š
        'gross_margin': 0.35,  # ç²—åˆ©ç‡35%
        'cost': 50 * 3000,  # åˆæœŸä»•å…¥1SKU3,000å††
        'description': 'æ–°å•†å“50SKUã‚’å°å…¥ï¼ˆæœˆå•†Â¥25ä¸‡æƒ³å®šï¼‰'
    }
]

print(f"\n\nğŸ”® ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³çµæœ")
print("="*120)

simulation_results = []

for scenario in scenarios:
    print(f"\n\n{'='*120}")
    print(f"ã€ã‚·ãƒŠãƒªã‚ªã€‘ {scenario['name']}")
    print(f"{'='*120}")
    print(f"   {scenario['description']}")
    
    # ã‚·ãƒŠãƒªã‚ªã”ã¨ã®è¨ˆç®—
    if 'price_change' in scenario:
        # ä¾¡æ ¼å¤‰æ›´ã‚·ãƒŠãƒªã‚ª
        affected_ratio = scenario.get('affected_sales_ratio', 1.0)
        price_change = scenario['price_change']
        quantity_change = scenario['quantity_elasticity']
        
        # æ–°ã—ã„å£²ä¸Š = (ä¾¡æ ¼ Ã— (1+ä¾¡æ ¼å¤‰æ›´ç‡)) Ã— (æ•°é‡ Ã— (1+æ•°é‡å¤‰æ›´ç‡))
        new_price = baseline_unit_price * (1 + price_change)
        new_quantity = baseline_daily_quantity * affected_ratio * (1 + quantity_change)
        affected_sales = new_price * new_quantity
        unaffected_sales = baseline_daily_sales * (1 - affected_ratio)
        new_daily_sales = affected_sales + unaffected_sales
        
        monthly_sales_change = (new_daily_sales - baseline_daily_sales) * 30
        
    elif 'staff_increase' in scenario:
        # ã‚¹ã‚¿ãƒƒãƒ•å¢—å“¡ã‚·ãƒŠãƒªã‚ª
        sales_increase = scenario['sales_increase']
        new_daily_sales = baseline_daily_sales * (1 + sales_increase)
        monthly_sales_change = (new_daily_sales - baseline_daily_sales) * 30
        
    elif 'inventory_increase' in scenario:
        # åœ¨åº«å¢—é‡ã‚·ãƒŠãƒªã‚ª
        opportunity_loss_reduction = scenario['opportunity_loss_current'] * scenario['stockout_reduction']
        monthly_sales_change = opportunity_loss_reduction
        new_daily_sales = baseline_daily_sales + (monthly_sales_change / 30)
        
    elif 'new_products' in scenario:
        # æ–°å•†å“å°å…¥ã‚·ãƒŠãƒªã‚ª
        new_product_sales = scenario['new_products'] * scenario['expected_sales_per_sku']
        monthly_sales_change = new_product_sales
        new_daily_sales = baseline_daily_sales + (monthly_sales_change / 30)
    
    # åˆ©ç›Šè¨ˆç®—
    gross_margin = scenario.get('gross_margin', 0.30)
    monthly_profit_change = monthly_sales_change * gross_margin
    monthly_cost = scenario['cost']
    net_profit = monthly_profit_change - monthly_cost
    roi = (net_profit / monthly_cost) * 100 if monthly_cost > 0 else 0
    
    # ãƒªã‚¹ã‚¯è©•ä¾¡
    risk_level = 'Low'
    if roi < 20:
        risk_level = 'High'
    elif roi < 50:
        risk_level = 'Medium'
    
    feasibility = 'High'
    if 'new_products' in scenario and scenario['new_products'] > 30:
        feasibility = 'Medium'
    if monthly_cost > 200000:
        feasibility = 'Medium' if feasibility == 'High' else 'Low'
    
    print(f"\nğŸ“Š äºˆæ¸¬çµæœ:")
    print(f"   ç¾çŠ¶æœˆå•†:     Â¥{baseline_daily_sales * 30:,.0f}")
    print(f"   äºˆæ¸¬æœˆå•†:     Â¥{new_daily_sales * 30:,.0f}")
    print(f"   å£²ä¸Šå¤‰åŒ–:     Â¥{monthly_sales_change:+,.0f} ({(monthly_sales_change/(baseline_daily_sales*30))*100:+.1f}%)")
    
    print(f"\nğŸ’° åç›Šæ€§:")
    print(f"   æŠ•è³‡é¡:       Â¥{monthly_cost:,.0f}")
    print(f"   å¢—åˆ†ç²—åˆ©:     Â¥{monthly_profit_change:,.0f} (ç²—åˆ©ç‡{gross_margin*100:.0f}%)")
    print(f"   ç´”åˆ©ç›Š:       Â¥{net_profit:,.0f}")
    print(f"   ROI:          {roi:+.1f}%")
    
    print(f"\nâš ï¸ ãƒªã‚¹ã‚¯ãƒ»å®Ÿç¾å¯èƒ½æ€§:")
    print(f"   ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«: {risk_level}")
    print(f"   å®Ÿç¾å¯èƒ½æ€§:   {feasibility}")
    
    print(f"\nğŸ’¡ æ¨å¥¨åˆ¤å®š:")
    if roi > 50 and risk_level in ['Low', 'Medium']:
        recommendation = 'âœ… å¼·ãæ¨å¥¨ - é«˜ROIã‹ã¤å®Ÿç¾å¯èƒ½'
        priority = 'High'
    elif roi > 20:
        recommendation = 'ğŸ“Š æ¨å¥¨ - å®Ÿæ–½ã‚’æ¤œè¨'
        priority = 'Medium'
    elif roi > 0:
        recommendation = 'âš ï¸ æ¡ä»¶ä»˜ã - æ…é‡ã«æ¤œè¨'
        priority = 'Low'
    else:
        pass
        recommendation = 'âŒ éæ¨å¥¨ - å®Ÿæ–½ã‚’è¦‹é€ã‚Š'
        priority = 'None'
    
    print(f"   {recommendation}")
    
    simulation_results.append({
        'ã‚·ãƒŠãƒªã‚ª': scenario['name'],
        'æŠ•è³‡é¡': monthly_cost,
        'å£²ä¸Šå¤‰åŒ–': monthly_sales_change,
        'ç´”åˆ©ç›Š': net_profit,
        'ROI': roi,
        'ãƒªã‚¹ã‚¯': risk_level,
        'å®Ÿç¾å¯èƒ½æ€§': feasibility,
        'å„ªå…ˆåº¦': priority
    })

# ç·åˆè©•ä¾¡ãƒ†ãƒ¼ãƒ–ãƒ«
print(f"\n\n{'='*120}")
print(f"ç·åˆè©•ä¾¡ãƒ»å„ªå…ˆé †ä½")
print(f"{'='*120}")

results_df = pd.DataFrame(simulation_results)
results_df = results_df.sort_values('ROI', ascending=False)

print(f"\n{'ã‚·ãƒŠãƒªã‚ª':<30} {'æŠ•è³‡é¡':>15} {'ç´”åˆ©ç›Š':>15} {'ROI':>10} {'ãƒªã‚¹ã‚¯':<10} {'å®Ÿç¾æ€§':<10} {'å„ªå…ˆåº¦'}")
print("="*120)

for _, row in results_df.iterrows():
    scenario_name = row['ã‚·ãƒŠãƒªã‚ª'][:28]
    cost = f"Â¥{row['æŠ•è³‡é¡']:,.0f}"
    profit = f"Â¥{row['ç´”åˆ©ç›Š']:,.0f}"
    roi_val = f"{row['ROI']:+.0f}%"
    risk = row['ãƒªã‚¹ã‚¯']
    feas = row['å®Ÿç¾å¯èƒ½æ€§']
    priority = row['å„ªå…ˆåº¦']
    
    print(f"{scenario_name:<30} {cost:>15} {profit:>15} {roi_val:>10} {risk:<10} {feas:<10} {priority}")

print(f"\nâœ… What-ifã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³å®Œäº†")
````
