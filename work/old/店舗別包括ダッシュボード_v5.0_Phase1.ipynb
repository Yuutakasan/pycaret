```python
# ğŸ” ãƒ‡ãƒ¼ã‚¿æ¤œè¨¼ï¼ˆå­˜åœ¨ãƒã‚§ãƒƒã‚¯ï¼‰

def validate_data_column(df, col_name, analysis_name="åˆ†æ"):
    """ãƒ‡ãƒ¼ã‚¿ã‚«ãƒ©ãƒ ã®å­˜åœ¨ã¨æœ‰åŠ¹æ€§ã‚’ãƒã‚§ãƒƒã‚¯"""
    if col_name not in df.columns:
        print(f"âš ï¸ {analysis_name}: '{col_name}' ã‚«ãƒ©ãƒ ãŒå­˜åœ¨ã—ã¾ã›ã‚“")
        return False
    
    non_null_count = df[col_name].notna().sum()
    coverage = non_null_count / len(df) * 100
    
    if coverage < 50:
        print(f"âš ï¸ {analysis_name}: '{col_name}' ã®ã‚«ãƒãƒ¬ãƒƒã‚¸ãŒä½ã„ ({coverage:.1f}%)")
        return False
    
    return True

print("\nğŸ” ãƒ‡ãƒ¼ã‚¿æ¤œè¨¼ä¸­...")
print("="*80)

# å¿…é ˆã‚«ãƒ©ãƒ ãƒã‚§ãƒƒã‚¯
required_cols = ['æ—¥ä»˜', 'å£²ä¸Šé‡‘é¡', 'åº—èˆ—']
for col in required_cols:
    if col in df_enriched.columns:
        print(f"âœ… å¿…é ˆã‚«ãƒ©ãƒ  '{col}' - å­˜åœ¨")
    else:
        print(f"âŒ å¿…é ˆã‚«ãƒ©ãƒ  '{col}' - ä¸è¶³")
        print(f"âŒ å¿…é ˆã‚«ãƒ©ãƒ  '{col}' - ä¸è¶³")

# ã‚ªãƒ—ã‚·ãƒ§ãƒ³ã‚«ãƒ©ãƒ ãƒã‚§ãƒƒã‚¯
optional_cols = {
    'æ°—è±¡ãƒ‡ãƒ¼ã‚¿': ['æœ€é«˜æ°—æ¸©', 'é™æ°´é‡'],
    'å‰å¹´ãƒ‡ãƒ¼ã‚¿': ['æ˜¨å¹´åŒæ—¥_å£²ä¸Š', 'æ˜¨å¹´åŒæ—¥_å®¢æ•°'],
    'æ™‚é–“å¸¯ãƒ‡ãƒ¼ã‚¿': ['æ™‚åˆ»', 'æ™‚é–“']
}

for category, cols in optional_cols.items():
    has_any = any(col in df_enriched.columns for col in cols)
    if has_any:
        available_cols = [col for col in cols if col in df_enriched.columns]
        print(f"âœ… {category}: {', '.join(available_cols)}")
    else:
        print(f"âŒ å¿…é ˆã‚«ãƒ©ãƒ  '{col}' - ä¸è¶³")
        print(f"âš ï¸ {category}: åˆ©ç”¨ä¸å¯ï¼ˆä»£æ›¿ãƒ­ã‚¸ãƒƒã‚¯ä½¿ç”¨ï¼‰")

print("="*80)
print("âœ… ãƒ‡ãƒ¼ã‚¿æ¤œè¨¼å®Œäº†\n")
```

```python
# ğŸ¯ åº—èˆ—é¸æŠ
my_df = df_enriched[df_enriched['åº—èˆ—'] == MY_STORE].copy()

# æ—¥æ¬¡é›†è¨ˆ
daily = my_df.groupby('æ—¥ä»˜').agg({
    'å£²ä¸Šé‡‘é¡': 'sum',
    'å£²ä¸Šæ•°é‡': 'sum',
    'å•†å“å': 'nunique',
    'æ˜¨å¹´åŒæ—¥_å£²ä¸Š': 'first'
}).reset_index()

daily['å‰å¹´æ¯”'] = (daily['å£²ä¸Šé‡‘é¡'] / daily['æ˜¨å¹´åŒæ—¥_å£²ä¸Š'] - 1) * 100
daily['å®¢å˜ä¾¡'] = daily['å£²ä¸Šé‡‘é¡'] / daily['å£²ä¸Šæ•°é‡']

# æœ€æ–°æ—¥
latest = daily.iloc[-1]
recent_7d = daily.tail(7)

# ã‚¨ã‚°ã‚¼ã‚¯ãƒ†ã‚£ãƒ–ã‚µãƒãƒªãƒ¼è¡¨ç¤º
print("\n" + "="*80)
print("ğŸ“Š ã‚¨ã‚°ã‚¼ã‚¯ãƒ†ã‚£ãƒ–ã‚µãƒãƒªãƒ¼".center(80))
print("="*80)

# 1. ä»Šæ—¥ã®å£²ä¸Šé€²æ—
target_sales = daily['å£²ä¸Šé‡‘é¡'].mean()
today_progress = (latest['å£²ä¸Šé‡‘é¡'] / target_sales) * 100
status_color = "ğŸŸ¢" if today_progress >= 95 else "ğŸŸ¡" if today_progress >= 85 else "ğŸ”´"

print(f"\n1ï¸âƒ£ æœ€æ–°æ—¥ã®å£²ä¸ŠçŠ¶æ³ ({latest['æ—¥ä»˜'].strftime('%Y-%m-%d')})")
print("-" * 80)
print(f"   å£²ä¸Šé‡‘é¡:  Â¥{latest['å£²ä¸Šé‡‘é¡']:,.0f} ({status_color} ç›®æ¨™æ¯” {today_progress:.1f}%)")
print(f"   å‰å¹´æ¯”:    {latest['å‰å¹´æ¯”']:+.1f}%")
print(f"   å®¢å˜ä¾¡:    Â¥{latest['å®¢å˜ä¾¡']:.0f}")

# 2. ç›´è¿‘7æ—¥é–“ã®ãƒˆãƒ¬ãƒ³ãƒ‰
trend_7d = recent_7d['å£²ä¸Šé‡‘é¡'].pct_change().mean() * 100
trend_emoji = "ğŸ“ˆ" if trend_7d > 0 else "ğŸ“‰"

print(f"\n2ï¸âƒ£ ç›´è¿‘7æ—¥é–“ã®ãƒˆãƒ¬ãƒ³ãƒ‰")
print("-" * 80)
print(f"   å¹³å‡æ—¥å•†:  Â¥{recent_7d['å£²ä¸Šé‡‘é¡'].mean():,.0f}")
print(f"   ãƒˆãƒ¬ãƒ³ãƒ‰:  {trend_emoji} {trend_7d:+.1f}% (å‰æ—¥æ¯”å¹³å‡)")
print(f"   å¹³å‡å‰å¹´æ¯”: {recent_7d['å‰å¹´æ¯”'].mean():+.1f}%")

# 3. ã‚¢ãƒ©ãƒ¼ãƒˆï¼ˆç•°å¸¸æ¤œçŸ¥ï¼‰
print(f"\n3ï¸âƒ£ ã‚¢ãƒ©ãƒ¼ãƒˆï¼ˆè¦æ³¨æ„äº‹é …ï¼‰")
print("-" * 80)

alerts = []
if latest['å‰å¹´æ¯”'] < -10:
    alerts.append(f"ğŸ”´ å‰å¹´æ¯”ãŒ{latest['å‰å¹´æ¯”']:.1f}%ã¨å¤§å¹…æ¸›å°‘ä¸­")
if trend_7d < -2:
    alerts.append(f"ğŸŸ¡ å£²ä¸ŠãŒä¸‹é™ãƒˆãƒ¬ãƒ³ãƒ‰ï¼ˆ7æ—¥å¹³å‡{trend_7d:.1f}%ï¼‰")
if latest['å®¢å˜ä¾¡'] < daily['å®¢å˜ä¾¡'].mean() * 0.9:
    alerts.append(f"ğŸŸ¡ å®¢å˜ä¾¡ãŒä½ä¸‹ä¸­ï¼ˆÂ¥{latest['å®¢å˜ä¾¡']:.0f} vs å¹³å‡Â¥{daily['å®¢å˜ä¾¡'].mean():.0f}ï¼‰")

if not alerts:
    print("   âœ… æ·±åˆ»ãªã‚¢ãƒ©ãƒ¼ãƒˆã¯ã‚ã‚Šã¾ã›ã‚“")
else:
    pass
    for alert in alerts:
        print(f"   {alert}")

# 4. ä»Šæ—¥ã®ã‚¢ã‚¯ã‚·ãƒ§ãƒ³ã‚¢ã‚¤ãƒ†ãƒ 
print(f"\n4ï¸âƒ£ ä»Šæ—¥ã®ã‚¢ã‚¯ã‚·ãƒ§ãƒ³ã‚¢ã‚¤ãƒ†ãƒ ")
print("-" * 80)
print("   â–¡ TOP10å•†å“ã®åœ¨åº«ç¢ºèª")
print("   â–¡ æ˜æ—¥ã®å¤©æ°—ã«å¿œã˜ãŸç™ºæ³¨èª¿æ•´")
print("   â–¡ å‰å¹´æ¯”ãƒã‚¤ãƒŠã‚¹å•†å“ã®è¦å› åˆ†æ")
print("   â–¡ ä»–åº—ã¨ã®æ¯”è¼ƒã§å¼±ç‚¹ã‚’ç¢ºèª")
print("   â–¡ æ¬ å“å•†å“ã®ãƒã‚§ãƒƒã‚¯")

print("\n" + "="*80)
```

# ğŸ”® ã€æ©Ÿèƒ½2ã€‘éœ€è¦äºˆæ¸¬ã‚¨ãƒ³ã‚¸ãƒ³ï¼ˆPyCaretçµ±åˆï¼‰

## æ˜æ—¥ãƒ»æ¥é€±ã®å£²ä¸Šã‚’äºˆæ¸¬ã—ã€ç™ºæ³¨é‡ã‚’è‡ªå‹•è¨ˆç®—

```python

# ğŸ“Š ã‚°ãƒ©ãƒ•ã®è¦‹æ–¹ã‚¬ã‚¤ãƒ‰
#
# ã€ç‰¹å¾´é‡é‡è¦åº¦ã‚°ãƒ©ãƒ•ã€‘
#   ãƒ»æ£’ãŒé•·ã„é …ç›® â†’ å£²ä¸Šäºˆæ¸¬ã«å¤§ããå½±éŸ¿ã™ã‚‹è¦ç´ 
#   ãƒ»ä¸Šä½3ã¤ã®è¦ç´ ã«æ³¨ç›®ã—ã¦æ–½ç­–ã‚’è€ƒãˆã‚‹
#
#   ä¾‹ï¼‰ã€Œæœ€é«˜æ°—æ¸©ã€ãŒä¸Šä½ â†’ æ°—æ¸©ã«ã‚ˆã‚‹å•†å“å…¥æ›¿ãŒåŠ¹æœçš„
#       ã€Œæ›œæ—¥ã€ãŒä¸Šä½ â†’ æ›œæ—¥åˆ¥ã®å“æƒãˆå¤‰æ›´ãŒé‡è¦
#       ã€Œæ˜¨å¹´åŒæ—¥_å£²ä¸Šã€ãŒä¸Šä½ â†’ å‰å¹´ãƒ‡ãƒ¼ã‚¿ã‚’å‚è€ƒã«ã—ãŸç™ºæ³¨ãŒæœ‰åŠ¹
#
#   âœ… åˆ¤æ–­åŸºæº–: é‡è¦åº¦0.1ä»¥ä¸Šã®è¦ç´ ã«é›†ä¸­ã—ã¦å¯¾ç­–ã‚’æ‰“ã¤


# ğŸ¤– PyCaretã«ã‚ˆã‚‹éœ€è¦äºˆæ¸¬ãƒ¢ãƒ‡ãƒ«æ§‹ç¯‰
print("\nğŸ¤– éœ€è¦äºˆæ¸¬ãƒ¢ãƒ‡ãƒ«ã‚’æ§‹ç¯‰ä¸­...")
print("   (åˆå›ã¯æ•°åˆ†ã‹ã‹ã‚‹å ´åˆãŒã‚ã‚Šã¾ã™)\n")

try:
    from pycaret.regression import *
    
    # ãƒ¢ãƒ‡ãƒªãƒ³ã‚°ç”¨ãƒ‡ãƒ¼ã‚¿æº–å‚™
    modeling_data = my_df.copy()
    
    # ç‰¹å¾´é‡é¸æŠï¼ˆãƒ¬ãƒ™ãƒ«2:ã‚¹ã‚¿ãƒ³ãƒ€ãƒ¼ãƒ‰ï¼‰
    feature_cols = [
        # æ™‚é–“åŸºæœ¬
        'æ›œæ—¥', 'æœˆ', 'æ—¥', 'é€±ç•ªå·',
        # æ™‚é–“ãƒ•ãƒ©ã‚°
        'ç¥æ—¥ãƒ•ãƒ©ã‚°', 'é€±æœ«ãƒ•ãƒ©ã‚°', 'å¹³æ—¥ãƒ•ãƒ©ã‚°',
        # ã‚¤ãƒ™ãƒ³ãƒˆ
        'çµ¦æ–™æ—¥', 'é€£ä¼‘ãƒ•ãƒ©ã‚°', 'é€£ä¼‘æ—¥æ•°', 'é€£ä¼‘åˆæ—¥', 'é€£ä¼‘æœ€çµ‚æ—¥',
        'GW', 'ç›†ä¼‘ã¿', 'å¹´æœ«å¹´å§‹',
        # å­¦æ ¡
        'å¤ä¼‘ã¿', 'å†¬ä¼‘ã¿',
        # å­£ç¯€å¤‰å‹•
        'å­£ç¯€å¤‰å‹•æŒ‡æ•°_æœˆ', 'å­£ç¯€å¤‰å‹•æŒ‡æ•°_é€±', 'å­£ç¯€_ãƒ”ãƒ¼ã‚¯æœŸ',
        # å‰å¹´æ¯”è¼ƒ
        'æ˜¨å¹´åŒæ—¥_å£²ä¸Š', 'æ˜¨å¹´åŒæ—¥_å®¢æ•°', 'æ˜¨å¹´åŒæ—¥_å®¢å˜ä¾¡',
        # å•†å“å±æ€§
        'ãƒ•ã‚§ã‚¤ã‚¹ããã‚Šå¤§åˆ†é¡'
    ]
    
    # æ°—è±¡ç‰¹å¾´é‡ï¼ˆåˆ©ç”¨å¯èƒ½ãªå ´åˆã®ã¿è¿½åŠ ï¼‰
    weather_cols = ['æœ€é«˜æ°—æ¸©', 'é™æ°´é‡', 'é™é›¨ãƒ•ãƒ©ã‚°', 'æœ€é«˜æ°—æ¸©_MA7', 'æ°—æ¸©ãƒˆãƒ¬ãƒ³ãƒ‰_7d']
    for col in weather_cols:
        if col in modeling_data.columns and modeling_data[col].notna().sum() > 0:
            feature_cols.append(col)
    
    # åˆ©ç”¨å¯èƒ½ãªåˆ—ã®ã¿é¸æŠ
    available_features = [col for col in feature_cols if col in modeling_data.columns]
    
    print(f"âœ… ä½¿ç”¨ç‰¹å¾´é‡: {len(available_features)}å€‹")
    print(f"   {', '.join(available_features[:10])}...")
    
    # æ—¥æ¬¡é›†è¨ˆï¼ˆå•†å“åˆ¥ï¼‰
    product_daily = modeling_data.groupby(['å•†å“å', 'æ—¥ä»˜']).agg({
        'å£²ä¸Šé‡‘é¡': 'sum',
        **{col: 'first' for col in available_features}
    }).reset_index()
    
    # æ¬ æå€¤å‰Šé™¤
    product_daily = product_daily.dropna(subset=['å£²ä¸Šé‡‘é¡'] + available_features)
    
    print(f"\nğŸ“Š ãƒ¢ãƒ‡ãƒªãƒ³ã‚°ç”¨ãƒ‡ãƒ¼ã‚¿:")
    print(f"   å•†å“æ•°: {product_daily['å•†å“å'].nunique()}")
    print(f"   ãƒ‡ãƒ¼ã‚¿æ•°: {len(product_daily):,}è¡Œ")
    
    if len(product_daily) >= 100:  # æœ€ä½100è¡Œå¿…è¦
        # PyCaretã‚»ãƒƒãƒˆã‚¢ãƒƒãƒ—
        print("\nâ³ PyCaretç’°å¢ƒã‚’ã‚»ãƒƒãƒˆã‚¢ãƒƒãƒ—ä¸­...")
        
        reg = setup(
            data=product_daily,
            ignore_features=['å•†å“å'],
            target='å£²ä¸Šé‡‘é¡',
            categorical_features=['ãƒ•ã‚§ã‚¤ã‚¹ããã‚Šå¤§åˆ†é¡'] if 'ãƒ•ã‚§ã‚¤ã‚¹ããã‚Šå¤§åˆ†é¡' in available_features else None,
            numeric_features=[col for col in available_features if col != 'ãƒ•ã‚§ã‚¤ã‚¹ããã‚Šå¤§åˆ†é¡'],
            fold_strategy='timeseries',
            fold=3,
            data_split_shuffle=False,
            fold_shuffle=False,
            normalize=True,
            remove_multicollinearity=True,
            multicollinearity_threshold=0.9,
            session_id=42,
            verbose=False,
            html=False
        )
        
        print("âœ… ã‚»ãƒƒãƒˆã‚¢ãƒƒãƒ—å®Œäº†")
        
        # ãƒ¢ãƒ‡ãƒ«æ¯”è¼ƒ
        print("\nâ³ æœ€é©ãƒ¢ãƒ‡ãƒ«ã‚’æ¢ç´¢ä¸­...")
        best_models = compare_models(
            include=['lightgbm', 'xgboost', 'rf', 'gbr'],
            n_select=1,
            sort='å¹³å‡çµ¶å¯¾èª¤å·®',
            verbose=False
        )
        
        best_model = best_models if not isinstance(best_models, list) else best_models[0]
        
        print(f"âœ… æœ€é©ãƒ¢ãƒ‡ãƒ«: {type(best_model).__name__}")
        
        # ãƒ¢ãƒ‡ãƒ«è©•ä¾¡
        results = pull()
        print(f"\nğŸ“ˆ ãƒ¢ãƒ‡ãƒ«æ€§èƒ½:")
        print(f"   MAE:  Â¥{results['å¹³å‡çµ¶å¯¾èª¤å·®'].mean():,.0f}")
        print(f"   RMSE: Â¥{results['äºŒä¹—å¹³å‡å¹³æ–¹æ ¹èª¤å·®'].mean():,.0f}")
        print(f"   R2:   {results['R2'].mean():.3f}")
        
        # ç‰¹å¾´é‡é‡è¦åº¦
        if hasattr(best_model, 'feature_importances_'):
            importance_df = pd.DataFrame({
                'ç‰¹å¾´é‡': get_config('X_train').columns,
                'é‡è¦åº¦': best_model.feature_importances_
            }).sort_values('é‡è¦åº¦', ascending=False)
            
            print(f"\nğŸ” ç‰¹å¾´é‡é‡è¦åº¦ TOP 10:")
            print("-" * 60)
            for idx, row in importance_df.head(10).iterrows():
                bar = "â–ˆ" * int(row['é‡è¦åº¦'] * 40)
                print(f"{row['ç‰¹å¾´é‡']:<30} {bar} {row['é‡è¦åº¦']:.4f}")
            
            # å¯è¦–åŒ–
            fig, ax = plt.subplots(figsize=(12, 6))
            importance_df.head(15).plot(x='ç‰¹å¾´é‡', y='é‡è¦åº¦', kind='barh', ax=ax, color='#4ECDC4')
            ax.set_title('å£²ä¸Šäºˆæ¸¬ã®ç‰¹å¾´é‡é‡è¦åº¦', fontsize=14, fontproperties=JP_FP)
            ax.set_xlabel('é‡è¦åº¦ã‚¹ã‚³ã‚¢', fontsize=12, fontproperties=JP_FP)
            ax.set_ylabel('', fontproperties=JP_FP)
            plt.tight_layout()
            plt.show()
        
        # äºˆæ¸¬ï¼ˆæ˜æ—¥åˆ†ï¼‰
        print(f"\nğŸ”® æ˜æ—¥ã®å£²ä¸Šã‚’äºˆæ¸¬ä¸­...")
        
        # æ˜æ—¥ã®ç‰¹å¾´é‡ã‚’æº–å‚™ï¼ˆæœ€æ–°æ—¥+1æ—¥ï¼‰
        tomorrow = product_daily['æ—¥ä»˜'].max() + timedelta(days=1)
        tomorrow_weekday = tomorrow.weekday()
        tomorrow_month = tomorrow.month
        
        print(f"   äºˆæ¸¬å¯¾è±¡æ—¥: {tomorrow.strftime('%Y-%m-%d (%a)')}")
        
        # ãƒ¢ãƒ‡ãƒ«ä¿å­˜ï¼ˆæ¬¡å›ç”¨ï¼‰
        save_model(best_model, 'demand_forecast_model')
        print(f"\nâœ… ãƒ¢ãƒ‡ãƒ«ã‚’ä¿å­˜ã—ã¾ã—ãŸ: demand_forecast_model.pkl")
        
    else:
        pass
        print(f"âš ï¸ ãƒ‡ãƒ¼ã‚¿ä¸è¶³: {len(product_daily)}è¡Œï¼ˆæœ€ä½100è¡Œå¿…è¦ï¼‰")
        print("   ã‚ˆã‚Šå¤šãã®ãƒ‡ãƒ¼ã‚¿ã‚’åé›†ã—ã¦ãã ã•ã„")
        
except ImportError:
    print("âš ï¸ PyCaretãŒã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã•ã‚Œã¦ã„ã¾ã›ã‚“")
    print("   ä»¥ä¸‹ã®ã‚³ãƒãƒ³ãƒ‰ã§ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã—ã¦ãã ã•ã„:")
    print("   pip install pycaret")
except Exception as e:
    print(f"âš ï¸ ã‚¨ãƒ©ãƒ¼: {str(e)}")
    print("   ãƒ‡ãƒ¼ã‚¿æ§‹é€ ã‚„ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã‚’ç¢ºèªã—ã¦ãã ã•ã„")
```
